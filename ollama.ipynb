```python
# 443の後ろに/を入れるとエラー(404)になる。
from langchain_community.llms import Ollama

llm = Ollama(base_url="https://epic-settling-mammoth.ngrok-free.app:443", model="gemma:7b")
llm.invoke("Who are you?")
```

```python
# 問題ない
from ollama import Client
client = Client(host='https://epic-settling-mammoth.ngrok-free.app:443/')

def chat():
  response = client.chat(model='gemma:7b', messages=[
    {
      'role': 'user',
      'content': '富士山について教えて',
    },
  ])
  print(response['message']['content'])

def chat_stream():
  stream_res = client.chat(model='gemma:7b', 
    messages=[
    {
      'role': 'user',
      'content': '富士山について教えて',
    }],
    stream=True,
  )
  for chunk in stream_res:
    print(chunk['message']['content'], end='', flush=True)

def embedding():
  response = client.embeddings(model='mxbai-embed-large', prompt='富士山について教えて')
  print(response['embedding'])
```
